```python
import numpy as np

# Simulate 100 coin tosses (0 = Tail, 1 = Head)
tosses = np.random.randint(0, 2, 100)

heads = np.sum(tosses == 1)
tails = np.sum(tosses == 0)

# Probabilities
prob_heads = heads / 100
prob_tails = tails / 100

print("Heads:", heads)
print("Tails:", tails)
print("Probability of Heads:", prob_heads)
print("Probability of Tails:", prob_tails)
```
