```python
import numpy as np

# Coefficient matrix
A = np.array([[2, 3],
              [4, 5]])

# Constant matrix
B = np.array([10, 20])

# Solve the system
solution = np.linalg.solve(A, B)

# Display results
print("Solution:")
print("x =", solution[0])
print("y =", solution[1])
```
