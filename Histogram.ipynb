```python
import numpy as np
import matplotlib.pyplot as plt

# Generate normal distribution data
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot histogram
plt.hist(data, bins=30)
plt.title("Normal Distribution Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
