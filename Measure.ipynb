```python
import numpy as np

# Sample dataset
data = np.array([10, 20, 30, 40, 50])

mean = np.mean(data)
median = np.median(data)
variance = np.var(data)
std_dev = np.std(data)

print("Mean:", mean)
print("Median:", median)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```
