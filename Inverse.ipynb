```python
import numpy as np

# Create a 3x3 matrix (must be non-singular)
matrix = np.array([[1, 2, 3],
                   [0, 1, 4],
                   [5, 6, 0]])

# Find inverse
inverse = np.linalg.inv(matrix)

# Display results
print("Matrix:\n", matrix)
print("Inverse:\n", inverse)
```
